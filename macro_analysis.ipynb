```python
from pynput import keyboard
import logging
from datetime import datetime

logging.basicConfig(filename='keyboard_input_logs.txt', level=logging.INFO, format='%(asctime)s - %(message)s')

def on_press(key):
    try:
        logging.info(f'Key pressed: {key.char}')
    except AttributeError:
        logging.info(f'Special key pressed: {key}')

def on_release(key):
    if key == keyboard.Key.esc:
        print('Exiting script...')
        # Stop listener on pressing the 'esc' key
        return False

print('Script is running. Press the "esc" key to exit.')

# Collect events until released
timestamps = []

def on_press_with_time(key):
    now = datetime.now()
    timestamp = now.strftime('%Y-%m-%d %H:%M:%S.%f')
    timestamps.append(timestamp)
    on_press(key)

with keyboard.Listener(on_press=on_press_with_time, on_release=on_release) as listener:
    listener.join()
```
